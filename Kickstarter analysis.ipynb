# Extracting data for Analysis-3

```python
data=pd.read_csv('ks-projects-201801.csv')
df=data[["main_category", "launched", "deadline", "state", "usd_goal_real", "backers"]].copy()
df['launched']=pd.to_datetime(data['launched'])
df['deadline']=pd.to_datetime(data['deadline'])

df['year']=df['launched'].dt.year
df['month']=df['launched'].dt.month_name().str.slice(stop=3)
df['day']=df['launched'].dt.day
df['day_of_week']=df['launched'].dt.weekday_name
# df['duration']=(df['deadline']-df['launched']).astype('timedelta64[D]').astype(int)
df.head()
```

```python
df['is_successful']=df['state'].apply(lambda x: 1 if x=='successful' else 0)
df['backers']=data['backers']
df.head()
```

```python
df3=pd.DataFrame()

df3['success_rate']=df.groupby('year')['is_successful'].mean()*100
df3['avg_backers']=df.groupby('year')['backers'].mean()
df3['year']=df3.index
display(df3)
```

```python
df3.to_csv("vis3-data.csv", index=False)
```
